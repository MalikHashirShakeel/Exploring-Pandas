```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df = pd.merge(df1, df2, on = "city")#Merge two dataframes with respect to a common column
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df = pd.merge(df1, df2, on = "city", how = "inner")#By default it is set to inner join
df
```

```python
df = pd.merge(df1, df2, on = "city", how = "outer")#Full outer join
df
```

```python
df = pd.merge(df1, df2, on = "city", how = "outer", indicator = True)#Full outer join with description of which data is from which table.
df
```

```python
df = pd.merge(df1, df2, on = "city", how = "left")#left outer join
df
```

```python
df = pd.merge(df1, df2, on = "city", how = "right")#right outer join
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df = pd.merge(df1, df2, on = "city")#Merge two dataframes with similar columns
df
```

```python
df = pd.merge(df1, df2, on = "city", how = "outer", suffixes = ("_left", "_right"))
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1.set_index("city", inplace = True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2, lsuffix = "_l", rsuffix = "_r")#Join df2 with df1
```
