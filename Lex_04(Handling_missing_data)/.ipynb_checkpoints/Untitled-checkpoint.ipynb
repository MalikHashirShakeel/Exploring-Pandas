```python
import pandas as pd
df = pd.read_csv("weather_data.csv", parse_dates = ["day"])#Treat each element of day column as datatype(day)
df
```

```python
type(df.day[0])#Get the type of the element of day column
```

```python
df.set_index("day", inplace = True)
df
```

```python
df_2 = df.fillna(0)#Replace all the not available values with 0
df_2
```

```python
df_2 = df.fillna({
    "temperature" : 0,
    "windspeed" : 0,
    "event" : "no record"
})#Specify the na values of individual columns
df_2
```

```python
df_3 = df.ffill()#Replace all the na values with previous cell
df_3
```

```python
df_3 = df.bfill()#Replace all the na values with the value in the next cell
df_3
```

```python
df_3 = df.ffill(axis = "columns")#Forward fill in the dimensions of columns
```

```python
df_3 = df.ffill(limit = 1)#Forward fill only once.
df_3
```

```python
df_4 = df.interpolate()#Replace the na values with the mean of the two adjacent values
df_4
```

```python
df_4 = df.interpolate(method = "time")#Replace the na values with the mean of the two adjacent values biased on the closer index
df_4
```

```python
df_5 = df.dropna()#Remove all the rows with na cells
df_5
```

```python
df_5 = df.dropna(how = "all")#Remove all the rows with na cells
df_5
```

```python
df_5 = df.dropna(thresh = 1)#Remove all the rows with na cells keeping those having atleast one non na value
df_5
```

```python
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)#Get the the missing dates in between
df
```
