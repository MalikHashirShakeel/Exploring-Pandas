```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data.csv")
df
```

```python
new_df = df.replace(-99999, np.NaN)#Replace all the -99999 with NaN
new_df
```

```python
new_df = df.replace({
    "temperature" : -99999,
    "windspeed" : -99999,
    "event" : "0"
}, np.NaN)#Replacce all the values in the given columns with NaN
new_df
```

```python
new_df = df.replace({
    -99999 : np.NaN,
    "0" : "No Event"
})#Replace all the given keys with given values
new_df
```

```python
new_df = df.replace({
    "temperature" : "[A-Za-z]",
    "windspeed" : "[A-Za-z]"
}, '', regex = True)#If any alphabet is detected in the above columns, replace them with ""
new_df
```

```python
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
new_df = df.replace(["exceptional", "average", "good", "poor"], [4, 3, 2, 1])#Replace all the given values with the corresponding values.
new_df
```
