```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
weather_df = pd.concat([india_weather, us_weather])#Concatinate the two dataframes by adding rows below
weather_df
```

```python
weather_df = pd.concat([india_weather, us_weather], ignore_index = True)#Concatinate the two dataframes by adding rows below without effecting the index
weather_df
```

```python
weather_df = pd.concat([india_weather, us_weather], keys = ["India", "Us"])#Concatinate the two dataframes by adding rows below With particular keys with them.
weather_df
```

```python
weather_df.loc["India"]#Locate the dataframe of key india
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df, windspeed_df], axis = 1)#Concatinate the two dataframes column wise
df
```

```python
windspeed_series = pd.Series([1, 2, 3], name = "Windspeed")
df = pd.concat([temperature_df, windspeed_series], axis = 1)#Concatinate the series with the dataframe.
df
```
